```python
import glob #to read the data files
import numpy as np #to math 
import math #to math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D 
import ipywidgets as widgets
from IPython.display import display
%matplotlib nbagg
```

```python
names=glob.glob("*.dat")

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

plt.suptitle("supernova")
    
def update_graph(i):

    ax.clear()
    
    #oppening the files
    file = open(names[i])
    lines = file.readlines()
    file.close()

    #temporary lists
    xe=[]
    ye=[]
    ze=[]

    #to separate the collumns
    for line in lines:
        if line.startswith('#'): continue
        co=line.rstrip().replace('INDEF','Nan').split()

        #saving in temporary lists
        xe.append(co[0])
        ye.append(co[1])
        ze.append(co[2])

    #to float
    xe = np.array(xe, dtype=float)
    ye = np.array(ye, dtype=float)
    ze = np.array(ze, dtype=float)


    ax.scatter(xe, ye, ze)
    print(names[i])
    ax.set_xlim3d(-20,50)
    ax.set_ylim3d(2000,9200)
    plt.show()
    
i = widgets.IntSlider(min= 0, max = len(names), value = 0, description = 'supernova:')
widgets.interactive(update_graph,i = i)
```
