```python
import numpy as np
import glob
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sn_files = glob.glob('*.dat')
sn_files
```

```python
time = []
wave = []
for i in range(0, len(sn_files)):
    
    time.append([])
    wave.append([])
```

```python
for i in range(0, len(sn_files)):
    
    file = open(sn_files[i])
    lines = file.readlines()
    file.close()
    
    time_temp = lines[0]
    time_temp = time_temp[:-1]
    time_temp = float(time_temp[1:])
    
    for line in lines:
        if line.startswith('#'): continue
        co=line.rstrip().replace('INDEF','Nan').split()
        
        time[i].append(time_temp)
        wave[i].append(float(co[0]))
```

```python
for i in range(0, len(time)):
    
    plt.plot(time[i], wave[i], 'k,')
    plt.xlim(-10, 50)
```
