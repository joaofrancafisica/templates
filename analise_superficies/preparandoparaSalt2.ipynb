```python
import numpy as np
import scipy as sp
import scipy.interpolate as interp
import math

file = open("realM0.dat")
lines = file.readlines()
file.close()
    
xsub = []
ysub = []
zsub = []
    
for line in lines:
    
    if line.startswith('#'): continue
    co=line.rstrip().replace('INDEF','Nan').split()

    xsub.append(co[0])
    ysub.append(co[1])
    zsub.append(co[2])

xsub = np.array(xsub, dtype=float)
ysub = np.array(ysub, dtype=float)
zsub = np.array(zsub, dtype=float)
```

```python
gridx = np.linspace(-20, 50, 50)
gridy = np.linspace(2000, 9200, 50)
B1, B2 = np.meshgrid(gridx, gridy, indexing='xy') 

fsub = interp.griddata((xsub,ysub),zsub,(B1,B2),method='linear')
```

```python
xx = []
yy = []
zz = []

for i in range(0,len(gridx)):
    for j in range(0, len(gridy)):
        
        xx.append(gridx[i])
        yy.append(gridy[j])
        
        if math.isnan(fsub[j][i]) == False:
            
            zz.append(fsub[j][i])
            
        else:
            
            zz.append(0)
```

```python
from tabulate import tabulate
table = []
for k in range(0,len(xx)):
    table.append((xx[k], yy[k], zz[k]))

f = open("newM0.dat", 'w')
f.write(tabulate(table, tablefmt="plain"))
f.close()
```
