```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
names = ["time", "wave", "m0", "m1"]
```

```python
# Reading SNEMO templates
snemo_df = pd.read_csv("/home/joao/Documents/pesquisa/BackupSurfaces/snemo/snemo2_ev.dat",
                       header=None,
                       sep='\s+',
                       names=names)
```

```python
# Reading PCA M0 template
pca_m0_df = pd.read_csv("/home/joao/Documents/pesquisa/BackupSurfaces/PCA3comp/M0.dat",
                       header=None,
                       sep='\s+',
                       names=names[0:3])
```

```python
# normalizing the data
snemo_df.m0 = snemo_df.m0/max(snemo_df.m0)
pca_m0_df.m0 = pca_m0_df.m0/max(pca_m0_df.m0)
```

```python
# Defining the grids
snemo_time_grid = snemo_df.time.unique()
snemo_wave_grid = snemo_df.wave.unique()

m0_time_grid = pca_m0_df.time.unique()
m0_wave_grid = pca_m0_df.wave.unique()

# number of grid numbers
time_number_snemo = len(snemo_time_grid)
wave_number_snemo = len(snemo_wave_grid)

time_number_m0 = len(m0_time_grid)
wave_number_m0 = len(m0_wave_grid)
```

```python
time_number_snemo
```

```python
# Reshaping the templates
snemo_reshape = np.array(snemo_df["m0"]).reshape(time_number_snemo, wave_number_snemo)
m0_reshape = np.array(pca_m0_df["m0"]).reshape(time_number_m0, wave_number_m0)
```

```python
import sncosmo
from sncosmo.salt2utils import BicubicInterpolator
```

```python
model = BicubicInterpolator(snemo_time_grid, snemo_wave_grid, snemo_reshape)
m0 = BicubicInterpolator(m0_time_grid, m0_wave_grid, m0_reshape)
```

```python
def disc(time_value, wave_value):
    """
    gives the discrepancy between a model and a test template
    """
    return np.abs(model(time_value, wave_value) - m0(time_value, wave_value))/model(time_value, wave_value)
```

```python
# plot grid
plot_grid_time = np.linspace(-10, 50, 61, dtype = int)
plot_grid_wave = np.linspace(3500, 8500, 501, dtype = int)
```

```python
# discrepancy values
comparison = disc(plot_grid_time, plot_grid_wave).T
```

```python
import matplotlib.cm as cm
import numpy as np
plt.rcParams.update({"font.size": 15})

plt.imshow(comparison, extent=(np.amin(plot_grid_time), np.amax(plot_grid_time), np.amin(plot_grid_wave), np.amax(plot_grid_wave)), origin='lower', cmap=cm.viridis, aspect='auto', interpolation = 'bilinear')
plt.colorbar()
plt.clim(0, 1)
plt.xlabel('tempo (dias)')
plt.ylabel(' $\lambda$ ($\AA$)')
plt.xlim(-10, 50)
plt.ylim(3500, 8000)
plt.subplots_adjust(bottom=0.15)
plt.subplots_adjust(left=0.2)
plt.savefig('/home/joao/Desktop/discrepancia_PCA_snemo.png', dpi = 200)
plt.show()
```
