```python
import scipy as sp
import scipy.interpolate
import numpy as np
import scipy.interpolate as interp
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

file = open("salt2_template_0.dat")
lines = file.readlines()
file.close()
    
xtemp = []
ytemp = []
ztemp = []
    
for line in lines:
    
    if line.startswith('#'): continue
    co=line.rstrip().replace('INDEF','Nan').split()

    xtemp.append(co[0])
    ytemp.append(co[1])
    ztemp.append(co[2])

xtemp = np.array(xtemp, dtype=float)
ytemp = np.array(ytemp, dtype=float)
ztemp = np.array(ztemp, dtype=float)
    
gradex,gradey = np.meshgrid(np.linspace(-20,50,70),np.linspace(2000,9200,50))

f = interp.griddata((xtemp,ytemp),ztemp,(gradex,gradey),method='linear')

%matplotlib notebook
fig = plt.figure(figsize=(10,6))
ax = axes3d.Axes3D(fig)
ax.set_ylim([2000,9200])
ax.set_xlim([-20,50])
ax.plot_wireframe(gradex, gradey, f)
ax.plot_surface(gradex, gradey, f,alpha=0.4)
```

```python
file = open("M0.dat")
lines = file.readlines()
file.close()
    
xsub = []
ysub = []
zsub = []
    
for line in lines:
    
    if line.startswith('#'): continue
    co=line.rstrip().replace('INDEF','Nan').split()

    xsub.append(co[0])
    ysub.append(co[1])
    zsub.append(co[2])

xsub = np.array(xsub, dtype=float)
ysub = np.array(ysub, dtype=float)
zsub = np.array(zsub, dtype=float)

gradex,gradey = np.meshgrid(np.linspace(min(xsub),max(xsub),100),np.linspace(min(ysub),max(ysub),100))

fsub = interp.griddata((xsub,ysub),zsub,(gradex,gradey),method='linear')

%matplotlib notebook
fig = plt.figure(figsize=(10,6))
ax = axes3d.Axes3D(fig)
ax.set_ylim([2000,9200])
ax.set_xlim([-20,50])
ax.plot_wireframe(gradex, gradey, fsub)
ax.plot_surface(gradex, gradey, fsub,alpha=0.4)
```

```python
from matplotlib import pyplot as plt
from matplotlib import animation
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = plt.figure()
ax = Axes3D(fig)
plt.xlabel('time (days)')
plt.ylabel('wavelength ($\AA$)')

def init():
    # Plot the surface.
    ax.plot_surface(gradex, gradey, f, cmap=cm.coolwarm,
                       linewidth=0, antialiased=True)
    return fig,

def animate(i):
    # azimuth angle : 0 deg to 360 deg
    ax.view_init(elev=10, azim=i*4)
    return fig,

ani = animation.FuncAnimation(fig, animate, init_func=init, frames=90, interval=50, blit=True)

fn = 'M0salt'

ani.save(fn+'.gif',writer='imagemagick',fps=1000/50)
```
